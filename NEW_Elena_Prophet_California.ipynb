```python
import warnings
import pandas as pd
from matplotlib import pyplot
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import r2_score, mean_squared_error
warnings.filterwarnings('ignore')
from statsmodels.tsa.arima.model import ARIMA
import statsmodels.api as sm
from pmdarima import auto_arima
from sklearn.model_selection import TimeSeriesSplit
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
from pmdarima import auto_arima
import pandas as pd
from sklearn.metrics import r2_score
import numpy as np
from prophet import Prophet
import pandas as pd
import itertools
from prophet import Prophet
from prophet.diagnostics import cross_validation, performance_metrics
import pandas as pd
```

```python
data = pd.read_csv('finalized_county.csv')
```

```python
california_data = data[data['state'] == 'California']
california_data['date'] = pd.to_datetime(california_data['date'], format="%Y-%m-%d")
california_data['county'].unique()
```

# Los Angeles

```python
la = california_data[california_data['county'] == 'Los Angeles'].loc[:, ['date', 'incremental_cases']]
la['incremental_cases'] = la['incremental_cases'].astype('int')
la['date'] = pd.to_datetime(la['date'], format="%Y-%m-%d")
```

```python
la.shape
```

```python
# use 85%-15% for train-test split
train = la.iloc[:230, :]
test = la.iloc[230:, :]
```

```python
max_date = train['date'].max()
min_date = train['date'].min()

max_date, min_date
```

```python
max_date = test['date'].max()
min_date = test['date'].min()

max_date, min_date
```

```python
# use non-differenced cases for test actuals
testing = data[data['county'] == 'Los Angeles'].loc[:, ['date', 'incremental_cases']]
test = testing.iloc[230:, :]

# Plotting actual test data
sns.lineplot(x="date", y="incremental_cases",legend = 'full' , data=testing)
```

```python
# Prepare the training data for Prophet
train = train.reset_index().rename(columns={'date': 'ds', 'incremental_cases': 'y'})
test = test.reset_index().rename(columns={'date': 'ds', 'incremental_cases': 'y'})

# Define the parameter grid to search
param_grid = {
    'changepoint_prior_scale': [0.001, 0.01, 0.1, 0.5, 1, 1.5, 2],
    'seasonality_prior_scale': [0.01, 0.1, 1.0, 1.5, 2],
    'seasonality_mode': ['multiplicative', 'additive'],
    'holidays_prior_scale':[0.01, 0.1, 1.0, 10.0]
}

# Generate all combinations of parameters
all_params = [dict(zip(param_grid.keys(), v)) for v in itertools.product(*param_grid.values())]
mses = []  # Store the MSEs for each params here

# Use cross-validation to evaluate all parameters
for params in all_params:
    m = Prophet(**params)
    m.fit(train)  # Fit model with given params
    df_cv = cross_validation(m, initial='190 days', period='180 days', horizon='20 days')
    df_p = performance_metrics(df_cv, rolling_window=1)
    mses.append(df_p['mse'].values[0])  # Use MSE

# Find the best parameters
tuning_results = pd.DataFrame(all_params)
tuning_results['mse'] = mses
best_parameters = tuning_results.loc[tuning_results['mse'].idxmin()]

print(tuning_results)
print(f"Best Parameters: {best_parameters}")
```

```python
from prophet import Prophet
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

train = train.reset_index().rename(columns={'date': 'ds', 'incremental_cases': 'y'})
test = test.reset_index().rename(columns={'date': 'ds', 'incremental_cases': 'y'})

# Your best parameters
best_params = {
    'changepoint_prior_scale': 2,
    'seasonality_prior_scale': 1.5,
    'seasonality_mode': 'additive',
    'holidays_prior_scale': 0.01,
}

# Create and fit a Prophet model with the best parameters
model = Prophet(changepoint_prior_scale=best_params['changepoint_prior_scale'],
                seasonality_prior_scale=best_params['seasonality_prior_scale'],
                seasonality_mode='additive')

model.fit(train)

# Make predictions on the training set
train_forecast = model.predict(train)

# Make predictions on the test set
test_forecast = model.predict(test)

# Calculate R^2 for the training set
train_r2 = r2_score(train['y'], train_forecast['yhat'])
print(f'Training Set R^2: {train_r2}')

# Calculate R^2 for the test set
test_r2 = r2_score(test['y'], test_forecast['yhat'])
print(f'Test Set R^2: {test_r2}')
```

```python
prediction = pd.DataFrame({'actual':test.y.reset_index(drop = True), 
              'pred':test_forecast.yhat.reset_index(drop = True)})
prediction['ds'] = test['ds'].reset_index(drop = True)
prediction.plot(x = 'ds', y = ['actual', 'pred'])
```

```python

# Plot the training data and predictions
fig, ax = plt.subplots(ncols=2, figsize=(14, 6))

# Configure date formatting and rotate date labels for better visibility
date_formatter = mdates.DateFormatter('%Y-%m-%d')
test['ds'] = pd.to_datetime(test['ds'])


# Training plot
ax[0].plot(train['ds'], train['y'], label='Actual')
ax[0].plot(train['ds'], train_forecast['yhat'], label='Predicted')
ax[0].fill_between(train['ds'], train_forecast['yhat_lower'], train_forecast['yhat_upper'], alpha=0.2)
ax[0].set_title('Prophet Los Angeles - Training Data and Predictions')
ax[0].xaxis.set_major_locator(mdates.AutoDateLocator())
ax[0].xaxis.set_major_formatter(date_formatter)
plt.setp(ax[0].get_xticklabels(), rotation=45, ha='right')

# Test plot
ax[1].plot(test['ds'], test['y'], label='Actual')
ax[1].plot(test['ds'], test_forecast['yhat'], label='Predicted')
ax[1].fill_between(test['ds'], test_forecast['yhat_lower'], test_forecast['yhat_upper'], alpha=0.2)
ax[1].set_title('Prophet Los Angeles - Test Data and Predictions')
ax[1].xaxis.set_major_locator(mdates.AutoDateLocator())
ax[1].xaxis.set_major_formatter(date_formatter)
plt.setp(ax[1].get_xticklabels(), rotation=45, ha='right')

# Adjust layout
plt.tight_layout()

# Show the plot with the new date format
plt.show()
```
