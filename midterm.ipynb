```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('finalized_county.csv').iloc[:, 1:]#.loc[:, ['date', 'entity', 'incremental_cases']]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

vis = data.loc[data['county'] == 'Los Angeles', ['date', 'cases', 'incremental_cases']]
vis['date'] = pd.to_datetime(vis['date'])

# Set the style of seaborn
sns.set_style("whitegrid")

# Create the line plot using seaborn
plt.figure(figsize=(9, 5))  # Adjust the figure size if needed
sns.lineplot(x='date', y='cases', data=vis, marker='o', color='b', linewidth=2)

# Rotate x-axis labels for better readability (optional)
plt.xticks(rotation=45)

# Add labels and title
plt.xlabel('Date', fontsize = 14)
plt.ylabel('Number of Cases', fontsize = 14)
plt.title('Cumulative COVID-19 Cases Over Time', fontsize = 16)

# Show plot
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
def incremental(df):
    df['incremental_cases'] = df['incremental_cases'].diff().fillna(0)
    return df
data = data.groupby('entity').apply(lambda x: incremental(x))
```

```python
from statsmodels.tsa.stattools import adfuller

def stationary_test(series):
    result = adfuller(series)
    return (result[0], result[1])
    
for i in list(data.entity.unique()):
    print(i, stationary_test(data[data['entity'] == i].incremental_cases))
```

```python
from statsmodels.tsa.stattools import adfuller
test = data[data['entity'] == 
    'California-Los Angeles'].incremental_cases
result = adfuller(test)

print('ADF Statistic:', result[0])
print('p-value:', result[1])
```

```python
from statsmodels.tsa.stattools import kpss

# Perform KPSS test
result = kpss(test, regression='c')

print('KPSS Statistic:', result[0])
print('p-value:', result[1])
print('Critical Values:')
for key, value in result[3].items():
    print(f'\t{key}: {value}')
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt
# Assuming 'data' is your time series data
vis = data[data['entity'] == 'Texas-Tarrant'].incremental_cases  # Your time series data

# Create a pandas Series from your data
ts = pd.Series(vis)

# Plot ACF
plt.figure(figsize=(20, 12))
plot_acf(ts, lags=30, alpha=0.05)
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.title('Autocorrelation Function (ACF)')
plt.show()

#Plot PACF
plt.figure(figsize=(12, 6))
plot_pacf(ts, lags=30, alpha=0.05)
plt.xlabel('Lag')
plt.ylabel('Partial Autocorrelation')
plt.title('Partial Autocorrelation Function (PACF)')
plt.show()
```
