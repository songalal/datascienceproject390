```python
import pandas as pd
df = pd.read_csv('variable_selected_county_dataset.csv', index_col=0)

for x in df.columns:
    print(x)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
s = sns.lineplot(data=df, x='date', y='incremental_cases', hue='entity')
s.set_xticks([df.date.min(), '2020-06-01', '2020-09-01', df.date.max()])
sns.move_legend(s, 'upper left', bbox_to_anchor=(1,1))
```

```python
s = sns.lineplot(data=df, x='date', y='mean_temp_3d_avg', hue='entity')
s.set_xticks([df.date.min(), '2020-06-01', '2020-09-01', df.date.max()])
sns.move_legend(s, 'upper left', bbox_to_anchor=(1,1))
```

```python
s = sns.lineplot(data=df, x='date', y='wind_speed', hue='entity')
s.set_xticks([df.date.min(), '2020-06-01', '2020-09-01', df.date.max()])
sns.move_legend(s, 'upper left', bbox_to_anchor=(1,1))
```

```python
s = sns.countplot(data=df, x='entity', hue='rain')
sns.move_legend(s, 'upper left', bbox_to_anchor=(1,1))
s.tick_params(axis='x',rotation=90)
#s.set_xticklabels(s.get_xticks(), size = 10)
```
