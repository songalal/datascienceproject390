# Focusing on columns 57:114

```python
data.columns[57:114]
```

```python
sub_data = data.iloc[:, 56:114]
sub_data['entity'] = data['state'] + '-' + data['county']
```

### Number of NAs for each column

```python
# number of NAs for each column
num_na_df = sub_data.isna().sum().reset_index().rename(
    columns = {'index':'col', 0:'num_na'}).sort_values(by = 'num_na', ascending = False)

# columns with NAs
num_na_df = num_na_df[num_na_df['num_na'] != 0]
num_na_df
```

### Checking missing values are stationary

```python
def stationary_check(df):
    non_stationary = []
    for i in df.entity.unique():
        entity_df = df[df['entity'] == i]
        for j in entity_df.columns.difference(['entity']):
            num_unique = len(entity_df[j].unique())
            if num_unique != 1:
                non_stationary.append([i, j])
    return non_stationary
```

```python
# columns with NA values
na_cols = num_na_df.col.to_list()

# checking whether columns with missing vals are stationary or time series
# use this to determine the 'location' of NA values
    # missing certain dates of a county
    # or missing for the county across dates
stationary_check(sub_data.loc[:, ['entity'] + na_cols])
```

^^ All of the missing variables are stationary

### Checking how many missing values are per county 

```python
# features with missing values are all stationary
df = sub_data.loc[:, ['entity'] + na_cols].drop_duplicates()

county_na = []
for i in df.entity:
    feature_df = df[df['entity'] == i]
    na_feature_num = feature_df.isna().sum(axis = 1).to_list()[0]
    county_na.append([i, na_feature_num])
```

```python
# number of missing features by entity
pd.set_option('display.max_rows', None)

pd.DataFrame(county_na, columns = ['entity', 'na_num']).sort_values(by = 'na_num', ascending = False)
```

### Drop those counties

```python
df = sub_data.loc[-sub_data['entity'].isin(['Virgin Islands-St. John', 'Virgin Islands-St. Croix', 
                       'Virgin Islands-St. Thomas']), ['entity'] + na_cols].drop_duplicates()
```

### Checking how many missing values are per feature after dropping some counties

```python
feature_na = []
for i in df.columns.difference(['entity']):
    feature_df = df.loc[:, ['entity', i]]
    na_num = len(feature_df[feature_df[i].isna()].entity.to_list())
    feature_na.append([i, na_num])
```

## EDA correlation   

```python
filtered_data = data.loc[-data['entity'].isin(['Virgin Islands-St. John', 'Virgin Islands-St. Croix', 'Virgin Islands-St. Thomas']), ['entity'] + na_cols]
```

```python
filtered_data
```

```python
# Convert the date column to datetime for proper filtering
data['date'] = pd.to_datetime(data['date'])
filtered_data['case_percentage'] = (data['cases'] / data['total_population']) * 100

# Filter the data for the date 2020-12-04
filtered_data = data[data['date'] == '2020-12-04']
filtered_data['case_percentage'] = (filtered_data['cases'] / filtered_data['total_population']) * 100
```

```python
import matplotlib.pyplot as plt

# Now let's create the scatter plot
plt.figure(figsize=(10, 6))

# Scatter plot for population vs cases
plt.scatter(filtered_data['total_population'], filtered_data['cases'])

# Adding title and labels
plt.title('Total Population vs Cases on 2020-12-04')
plt.xlabel('Total Population')
plt.ylabel('Cases')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Now let's create the scatter plot
plt.figure(figsize=(10, 6))

# Scatter plot for population vs cases
plt.scatter(filtered_data['case_percentage'], filtered_data['percent_vaccinated'])

# Adding title and labels
plt.title('Percent Vaccinated vs Case over Population Percentage')
plt.xlabel('Case over Population Percentage')
plt.ylabel('Percent Vaccinated')

# Show the plot
plt.show()
```
