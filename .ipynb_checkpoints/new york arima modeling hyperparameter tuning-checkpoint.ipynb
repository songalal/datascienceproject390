```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA

#85 percent train 15 percent test
```

```python
df = pd.read_csv('variable_selected_county_dataset.csv')
```

```python
df.entity.unique()
```

```python
df.isna().sum()
```

```python
# get selected counties
toWorkOn = df.loc[(df.county =='Maricopa') | (df.county == 'Los Angeles') | (df.county == 'Santa Clara')]
```

```python
maricopa = toWorkOn[toWorkOn.county == 'Maricopa']
maricopa = maricopa[['date', 'cases']]
maricopa.set_index('date',inplace=True)
```

```python
#maricopa.plot(x='date', y='cases')
maricopa['Change'] = maricopa['cases'].diff()
maricopa.drop(columns=['cases'], inplace=True)
maricopa = maricopa.iloc[1: , :]
```

```python
maricopa
```

```python
from matplotlib import pyplot as plt

train = maricopa.head(int(0.85*len(maricopa)))
test = maricopa.tail(int(0.15*len(maricopa)))
model = ARIMA(train, order=(5,1,0))
model_fit = model.fit()
# summary of fit model
print(model_fit.summary())
# line plot of residuals
residuals = pd.DataFrame(model_fit.resid)
residuals.plot()
plt.show()
# density plot of residuals
residuals.plot(kind='kde')
plt.show()
# summary stats of residuals
print(residuals.describe())
```

```python
foo = pd.DataFrame(model_fit.resid)
foo.columns=['Change']
predictedTrain = train + foo
predictedTrain
```

```python
predictedTrain.plot(y='Change')
```

```python
train.plot(y='Change')
```

```python
forecast = model_fit.forecast(steps=47)

predictTest = pd.DataFrame(forecast)
predictTest.columns=['Change']
predictTest
```

```python
predictTest.plot(y='Change')
```

```python
test.plot(y='Change')
```
