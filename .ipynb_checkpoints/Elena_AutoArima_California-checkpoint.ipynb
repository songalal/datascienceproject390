# AutoArima Models for California Counties 

```python
import warnings
import pandas as pd
from matplotlib import pyplot
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import r2_score, mean_squared_error
warnings.filterwarnings('ignore')
from statsmodels.tsa.arima.model import ARIMA
import statsmodels.api as sm
import os
from pmdarima import auto_arima
```

```python
data = pd.read_csv('finalized_county.csv')
```

```python
california_data = data[data['state'] == 'California']
california_data['date'] = pd.to_datetime(california_data['date'], format="%Y-%m-%d")
california_data['county'].unique()
```

## Los Angeles

```python
la = california_data[california_data['county'] == 'Los Angeles'].loc[:, ['date', 'incremental_cases']]

def stationary_test(series):
    result = adfuller(series)
    return (result[0], result[1])

# LA data is non-stationary without any differencing
stationary_test(la.incremental_cases)

# test statistic (hypothesis that non-stationarity cannot be rejected)
# p-value is greater than 0.05 
```

```python
# first-level differencing
la['incremental_cases'] = la['incremental_cases'].diff().fillna(la['incremental_cases']).astype('int')

# la is still non-stationary after first-level differencing
stationary_test(la.incremental_cases)

# test statistic (hypothesis that non-stationarity cannot be rejected)
# p-value is greater than 0.05 
```

```python
# second-level differencing
la['incremental_cases'] = la['incremental_cases'].diff().fillna(la['incremental_cases']).astype('int')

# la is stationary after second-level differencing
stationary_test(la.incremental_cases)

# test statistic (hypothesis that non-stationarity can be rejected)
# p-value is much less than 0.05
```

```python
# using differenced cases for train
train = la.iloc[:230, :]
```

```python
# use non-differenced cases for test actuals
testing = data[data['county'] == 'Los Angeles'].loc[:, ['date', 'incremental_cases']]
test = testing.iloc[230:, :]

sns.lineplot(x="date", y="incremental_cases",legend = 'full' , data=testing)
```

### Baseline Model

```python
model = auto_arima(train['incremental_cases'], seasonal=False, d=0, suppress_warnings=True, stepwise=True)

print(model.summary())

# Forecast for the number of periods equal to the test set size
n_periods = test.shape[0]
forecast, conf_int = model.predict(n_periods=n_periods, return_conf_int=True)

# Print the forecast and confidence interval
print("Forecast:", forecast)
print("Confidence Interval:", conf_int)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Plot historical data

plt.plot(la['date'], la['incremental_cases'], label='Actual')

# Plot the forecasted values
plt.plot(test['date'], forecast, color='orange', label='Forecast')

# Calculate and plot the confidence intervals
lower_bounds = conf_int[:, 0]
upper_bounds = conf_int[:, 1]
plt.fill_between(test['date'], lower_bounds, upper_bounds, color='k', alpha=0.2, label='Confidence Interval')

# Final plot formatting
plt.title('AutoARIMA Forecast with Confidence Interval')
plt.xlabel('Date')
plt.ylabel('Incremental Cases')
plt.legend()
plt.show()
```

```python
from pmdarima import auto_arima
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
import pandas as pd

# Assuming 'train' and 'test' DataFrames are already defined and include 'incremental_cases' column
model = auto_arima(train['incremental_cases'], seasonal=False, d=0, suppress_warnings=True, stepwise=True)

print(model.summary())

# Forecast for the number of periods equal to the test set size
n_periods = test.shape[0]
forecast, conf_int = model.predict(n_periods=n_periods, return_conf_int=True)

# Manually apply differencing logic similar to your original ARIMA example
# Assuming the initial differencing was meant to adjust the series level, then reverse it after forecasting
# Initial manual differencing reversal (simulating your ARIMA process)
forecast_diff_reversed = pd.concat([pd.Series([train['incremental_cases'].iloc[-1]]), pd.Series(forecast)]).cumsum()[1:]

# Print the adjusted forecast and confidence interval
print("Adjusted Forecast:", forecast_diff_reversed)
print("Confidence Interval:", conf_int)

# Plot historical data
plt.plot(train['date'], train['incremental_cases'], label='Actual')

# Plot the forecasted values, adjust the dates for plotting if necessary
plt.plot(test['date'], forecast_diff_reversed, color='orange', label='Forecast')

# Calculate and plot the confidence intervals, adjust if you apply any transformation to the forecast
lower_bounds = conf_int[:, 0]
upper_bounds = conf_int[:, 1]
plt.fill_between(test['date'], lower_bounds, upper_bounds, color='k', alpha=0.2, label='Confidence Interval')

# Final plot formatting
plt.title('AutoARIMA Forecast with Confidence Interval')
plt.xlabel('Date')
plt.ylabel('Incremental Cases')
plt.legend()
plt.show()

# Calculate R-squared for the adjusted forecast
r2 = r2_score(test['incremental_cases'], forecast_diff_reversed)
print("R-squared", r2)
```

```python
test['date']
```

### Tuning Hyperparameters

```python
p_values = [0, 1, 2, 3]
d_values = [0, 1, 2, 3]
q_values = [0, 1, 2, 3]

P_values = [0, 1, 2, 3]
D_values = [0, 1]
Q_values = [0, 1, 2, 3]
Season_length = [7]  # Assuming a weekly seasonality

prev_mse = float('inf')
best_params = (None, None)  # Initialized to a safer default

for p in p_values:
    for d in d_values:
        for q in q_values:
            for P in P_values:
                for D in D_values:
                    for Q in Q_values:
                        for S in Season_length:
                            try:
                                model = SARIMAX(train.incremental_cases,
                                                order=(p, d, q),
                                                seasonal_order=(P, D, Q, S),
                                                enforce_stationarity=False,
                                                enforce_invertibility=False).fit(disp=0)

                                pred = model.forecast(steps=40)
                                pred = pd.concat([pd.Series([-62]), pred]).cumsum()[1:]
                                pred = pd.concat([pd.Series([101]), pred]).cumsum()[1:]

                                mse = mean_squared_error(test.incremental_cases, pred)

                                if mse < prev_mse:
                                    prev_mse = mse
                                    best_params = ((p, d, q), (P, D, Q, S))

                                print(f"MSE: {mse} for parameters (p, d, q): {(p, d, q)} and (P, D, Q, S): {(P, D, Q, S)}")
                            except Exception as e:
                                print(f"Failed to fit model for parameters (p, d, q): {(p, d, q)}, (P, D, Q, S): {(P, D, Q, S)}. Error: {e}")

if best_params[0] is not None:
    print(f"Best Parameters: Non-seasonal (p, d, q): {best_params[0]}, Seasonal (P, D, Q, S): {best_params[1]} with MSE: {prev_mse}")
else:
    print("No successful model fit was found.")
```

```python
model = model = sm.tsa.statespace.SARIMAX(train.incremental_cases, order=(1, 0, 0), seasonal_order=(0,0,2,7)).fit()
pred = pd.concat([pd.Series([-62]), model.forecast(steps=40)]).cumsum()[1:]
pred = pd.concat([pd.Series([101]), pred]).cumsum()[1:]
r2 = r2_score(test.incremental_cases, pred)
r2
```

```python
prediction = pd.DataFrame({'actual':test.incremental_cases.reset_index(drop = True), 
              'pred':pred.reset_index(drop = True)})
prediction['date'] = test['date'].reset_index(drop = True)
prediction.plot(x = 'date', y = ['actual', 'pred'])
```
