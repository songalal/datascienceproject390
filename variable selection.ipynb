```python
Xvars = df.columns
toCheck = []
for x in Xvars:
    if x != 'date' and x != 'county' and x != 'state' and x != 'cases' and x != 'deaths' and x != 'entity' and x != 'incremental_cases':
        toCheck.append(x)
toCheck
```

```python
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']

newdf = df.select_dtypes(include=numerics)
newdf.drop(columns = ['cases', 'deaths'], inplace=True)
```

```python
print(newdf.corr().incremental_cases)
```

```python
newdf.corr().incremental_cases.index[0]

for x in newdf.corr().incremental_cases.index:
    print(x, newdf.corr().incremental_cases.at[x])
```

```python
#found variables that were numbers and percentage of population. anything population related should convert to per-capita to avoid over bias w/
#population, so this code ids the ones to remove and the ones to convert

toRemove = []
toConvert = []
for x in newdf.columns:
    if 'num_' in x:
        checkPerc = x
        checkPerc = 'percent_' + checkPerc[4:]
        if checkPerc in newdf.columns:
            print(x, checkPerc)
            toRemove.append(x)
        else:
            print(x, 'NONE FOUND')
            toConvert.append(x)
```

```python
toConvert
```

```python
#drops the columns
df.drop(columns=toRemove, inplace=True)
```

```python
#for the other columns without a corresponding percentage, will convert them by dividing the column by population of the county
for index, row in df.iterrows():
    divisor = row['population']  # Divisor from column B
    for col in toConvert:
        df.at[index, col] = row[col] / divisor
```
